# type codes available for the array module for various data types.

1. i - integert type (2 bytes)
2. f - float type (4 bytes)
3. d - float type (8 bytes)
4. u - unicode character (2 bytes)

```python
# creating an array

a = arr.array('i', [1,2,3,4,5])
b = arr.array('f', [1.0, 2.6, 2.7, 1.9, 5.1])
c = arr.array('d', [10.4, 3.7, 4.9, 1.9])
d = arr.array('u', ['a', 'b', 'c', 'd'])

print(a.typecode)
print(b.typecode)
print(c.typecode)
print(d.typecode)

print(a)
print(b)
print(c)
print(d)
```
